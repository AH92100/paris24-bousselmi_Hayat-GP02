```python
import pandas as pd
import plotly.express as px
```

```python
# Read birth rate data from Excel file
df = pd.read_excel('birthRateData.xlsx')

# Melt the dataframe to transform the year columns into rows
df_melted = pd.melt(df, id_vars='Country Name', var_name='Year', value_name='Birth Rate')

# Convert 'Year' to int
df_melted['Year'] = df_melted['Year'].astype(int)

# Extracting data from 1970 to 2021
start_date = 1970
end_date = 2021
df_melted = df_melted[(df_melted['Year'] >= start_date) & (df_melted['Year'] <= end_date)]

# Prepare data for animation
counts = df_melted.groupby(['Year', 'Country Name']).sum().reset_index()

# Create the animation of the map
birthMap = px.choropleth(counts, locations='Country Name', locationmode="country names", hover_name='Country Name', 
                    animation_frame='Year', color='Birth Rate', color_continuous_scale=px.colors.sequential.Plasma)

# Update the layout, title and geo
birthMap.update_layout(title_text='Birth rate per year', title_x=0.5, geo=dict(showframe=False, showcoastlines=False))

# Write to HTML and display, automatically open a window on the browser too
birthMap.write_html('Birth rate map.html', auto_open=True)
birthMap
```
